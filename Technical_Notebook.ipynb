```python
train_dir = 'train'
val_dir = 'valid'
```

```python
classes = os.listdir(train_dir)
print(classes)
```

```python
for i in classes:
    print(i)
```

```python
plant_count = {}

for plant in classes:
    plant_count[plant] = len(os.listdir(train_dir + '/' + plant))

total_img = pd.DataFrame(plant_count.values(), index = plant_count.keys(), columns = ['No of Images'])

total_img['Percentage of Data'] = total_img['No of Images'] / sum(total_img['No of Images'])
```

```python
total_img
```

```python
plant_types = []
disease_plant = 0

for plant,num in plant_count.items():
    if plant.split('___')[0] not in plant_types:
        plant_types.append(plant.split('___')[0])
    if plant.split('___')[1] != 'healthy':
        disease_plant +=1
```

```python
print(plant_types)
print('Healthy Plants:', len(plant_types))
print('Disease Plants:', disease_plant)
```

```python
import os

# IMPORTANT: PATH MIGHT BE DIFFERENT. SEE STEP 6
os.environ["KERAS_BACKEND"] = "plaidml.keras.backend"
# os.environ["RUNFILES_DIR"] = "/Library/Frameworks/Python.framework/Versions/3.7/share/plaidml"
# os.environ["PLAIDML_NATIVE_PATH"] = "/Library/Frameworks/Python.framework/Versions/3.7/lib/libplaidml.dylib"

import keras 
from keras import backend as K
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import EarlyStopping
from keras.losses import categorical_crossentropy

train_data_generator = ImageDataGenerator(rescale = 1./255,
                                         zoom_range = 0.2,
                                         shear_range = 0.2,
                                         width_shift_range = 0.2,
                                         height_shift_range = 0.2,
                                         horizontal_flip = True)

valid_data_generator = ImageDataGenerator(rescale = 1./255)

train_data = train_data_generator.flow_from_directory(train_dir, 
                                                     target_size = (224,224),
                                                     batch_size = 32,
                                                     class_mode = 'categorical',
                                                     seed=42)

valid_data = valid_data_generator.flow_from_directory(val_dir,
                                                     target_size = (224,224))

base_model = keras.applications.VGG19(weights='imagenet', include_top = False, input_tensor = keras.Input(shape = (224,224,3)))

base_model.trainable = False

model = keras.models.Sequential([base_model,
                                keras.layers.GlobalAveragePooling2D(),
                                keras.layers.Dense(128, activation = 'relu'),
                                keras.layers.Dropout(0.2),
                                keras.layers.Dense(38, activation = 'softmax')])

model.compile(optimizer = keras.optimizers.Adam(),
             loss = categorical_crossentropy,
             metrics = ['accuracy'])

history = model.fit_generator(train_data, 
                   steps_per_epoch = 150,
                   validation_data = valid_data,
                   epochs = 100,
                   validation_steps = 100,
                   callbacks=[EarlyStopping(patience=5)])
```
